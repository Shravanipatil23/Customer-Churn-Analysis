### Importing required library for futher ML transformations and predictions

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df_2 = df.copy()
```

```python
label_encoder = LabelEncoder()
df_2['Gender'] = label_encoder.fit_transform(df_2['Gender'])
df_2['Promotion_Response'] = label_encoder.fit_transform(df_2['Promotion_Response'])
```

```python
df_2.head()
```

```python
X = df_2.drop(['Customer_ID','Target_Churn'], axis = 1)
y = df_2['Target_Churn']
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
y.head(), X_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

```python
X_train.head(10), y_train.head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
rf_model = RandomForestClassifier(random_state = 42)
rf_model.fit(X_train, y_train)
y_pred = (rf_model.predict(X_test))
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
classification_rep
```

```python
# Prepare predictions for export to CSV
predictions_df = X_test.copy()
predictions_df['Actual_Churn'] = y_test
predictions_df['Predicted_Churn'] = y_pred
predictions_df['Customer_id'] = df['Customer_ID']
# Save the predictions to a CSV file
predictions_df.to_csv('churn_predictions.csv', index=False)
```
